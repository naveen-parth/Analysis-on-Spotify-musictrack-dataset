```python
most_popular = df_tracks.query('popularity>80',inplace= False).sort_values('popularity', ascending=False)
most_popular[::15]
```

```python
df_tracks.set_index("release_date", inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks["duration_in_min"]= df_tracks["duration_ms"].apply(lambda x: round(x/60000))
df_tracks.replace("duration_ms","duration_in_min",inplace = True)
df_tracks.head()
```

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
sns.regplot(data = sample_df, y = "loudness", x = "energy", color = "c").set(title="Loudness vs Energy correlation")
```

```python
sns.regplot(data=sample_df, y = "popularity", x="acousticness", color = "b").set(title="popularity vs acousticness correlation")
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True, aspect=2, height=6, kind="hist").set(title="Number of songs per year")
```

```python
total_dr=df_tracks.duration_ms
```

```python
sns.lineplot(x=years,y=total_dr,data=df_tracks).set(title="Year vs Duration")
plt.show()
```
